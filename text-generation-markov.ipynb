# Text Generation using a Markov Chain

We'll be using code from this repository on [Github](https://github.com/jsvine/markovify).

### Import the package

```python
!pip install markovify
import markovify
```

### Pull in the raw text

```python
with open("sonnets.txt") as f:
    text = f.read()
```

### Build the model

```python
text_model = markovify.Text(text)
```

### Print 5 sentences generated by the model

```python
for i in range(5):
    print(text_model.make_sentence())
```

### Print three generated sentences of no more than 140 characters

```python
# Print three randomly-generated sentences of no more than 140 characters
for i in range(3):
    print(text_model.make_short_sentence(140))
```
